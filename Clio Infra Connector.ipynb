```python
import urllib2 
import simplejson
import json
import pandas as pd
import re
import vincent
from pandas.io.json import json_normalize

def load_api_data(apiurl, fileID):
    jsondataurl = apiurl
    
    req = urllib2.Request(jsondataurl)
    opener = urllib2.build_opener()
    f = opener.open(req)
    dataframe = simplejson.load(f, "utf-8")
    return dataframe

jsonapi = "http://clearance.sandbox.socialhistoryservices.org/collabs/static/data/dataframe27_0.json"
dataframe = load_api_data(jsonapi, '')

clioframe = pd.DataFrame(dataframe)
clioframe.head()
```

```python
df = clioframe.transpose()

standards = df.drop(df.columns[[0]], axis=1)
countries = clioframe[clioframe.columns[1]]
codes = clioframe[clioframe.columns[0]]
c = pd.DataFrame({'codes': codes, 'countries': countries})
oecd = c[c['codes'] > 0]
oecd
```
