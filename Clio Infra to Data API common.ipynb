```python
import urllib2 
import simplejson
import json
import pandas as pd
import re
import vincent
import numpy as np
from pandas.io.json import json_normalize

def load_api_data(apiurl, fileID):
    jsondataurl = apiurl
    
    req = urllib2.Request(jsondataurl)
    opener = urllib2.build_opener()
    f = opener.open(req)
    dataframe = simplejson.load(f, "utf-8")
    return dataframe

def json_dict(sqlnames, jsondataname, data):
        jsonlist = []
        jsonhash = {}

        for valuestr in data:
            datakeys = {}
            for i in range(len(valuestr)):
               name = sqlnames[i]
               value = valuestr[i]
               datakeys[name] = value
               #print "%s %s", (name, value)
            jsonlist.append(datakeys)

        jsonhash[jsondataname] = jsonlist;
        json_string = json.dumps(jsonhash, encoding="utf-8", sort_keys=True, indent=4)

        return json_string

def loadcodes(dataframe):
    clioframe = pd.DataFrame(dataframe)
    countries = clioframe[clioframe.columns[1]]
    codes = clioframe[clioframe.columns[0]]

    code2country = {}
    for i in range(len(codes)):
        code = codes[i]
        country = countries[i]
        code2country[code] = country
        
    return code2country
    
def loadlexicon(header):
    lexicon = {}
    for x in sorted(header):
        lexicon[x] = header[x]        
    return lexicon

def nlgis_engine(lex, indicator, loccodes, dataframe, customyear, DEBUG):
    yearline = dataframe[1]
    allyears = []
    nlgis_frame = []
    yearsstat = {}
    
    (years, vocab) = ({}, {})
    for line in yearline:
        try:
            if not customyear:
                years[line] = int(yearline[line])
            elif int(customyear) == int(yearline[line]):
                years[line] = int(yearline[line])
            allyears.append(int(yearline[line]))
                            
        except:
            if not indicator:
                for item in lex:
                    if lex[item] == 'Code':
                        indicator = item
                vocab[line] = yearline[line]

    counter = 0
    for line in dataframe:
        (country, itemID) = ('', 0)
        order = []
        
        for item in line:
            row = []
            value = line[item]
            field = lex[item]

            if not country:
                if field == 'Continent, Region, Country':
                    country = value
            try:
                thisyear = years[item]
                if int(field):
                    thisyear = int(field)
                    if value > 0:
                        print str(indicator) + ' ' + str(thisyear) + ' ' + str(country) + ' ' + str(value) + '\n'
                        row.append(indicator)
                        row.append(loccodes[country])
                        row.append(thisyear)
                        row.append(country)
                        row.append(value)
                        row.append(counter)
                        counter = counter + 1
                        
                        #intyear = int(thisyear)
                        #if intyear in yearsstat:
                        #    yearsstat[intyear] = yearsstat[intyear] + 1
                        #else:
                        #    yearsstat[intyear] = 1

            except:
                do = 1
                if DEBUG:
                    print line[item]
            itemID = itemID + 1
            if row:
                nlgis_frame.append(row)
    
    return (nlgis_frame, allyears, yearsstat)

def liveapi(root, apipoint, year):
    jsonapi = root + "/collabs/static/data/" + apipoint
    dataframe = load_api_data(jsonapi, '')
    orderline = dataframe[1]
    lex = loadlexicon(orderline)    
    loccodes = loadcodes(dataframe)
    DEBUG = 0
    indicator = ''
    (frame, years, stats) = nlgis_engine(lex, indicator, loccodes, dataframe, year, DEBUG)
    names = ['indicator', 'country', 'year', 'intcode', 'value', 'id']
    jsondata = json_dict(names, 'data', frame)
    return (jsondata, years, stats)
    
```
